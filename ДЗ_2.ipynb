```python
from math import factorial
import numpy as np
```

```python
def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
def poisson_proba(k, lambda_):
    
    return (lambda_ ** k) * (np.exp(-lambda_)) / np.math.factorial(k)
```

**Задача 2**

Вероятность поймать рыбу при одном забрасывании удочки равна 0.01. Какова вероятность того, что число пойманных рыб при 200 забросах удочки будет не менее 5 и не более 10?

*т.к. в задача с счетчиком (число пойманных рыб) значит решаем задачу по формуле Пуассона*
$$P(X = k) = \dfrac{a^k e ^{-a}}{k!}.$$

```python
а = 200 * 0.01
```

```python
lambda_ = 2.0
```

*найдем вероятность выловить не менее 5 рыб, т.е. от 6 и до бесконечности:*

$$P(X > 5) = 1 - P(X \leq 5)$$

```python
P1 = 1 - sum(poisson_proba(k=i, lambda_=lambda_) for i in range(6))
print(P1)
```

*число пойманных рыб не более 10 означает, что нужно сложить несколько вероятностей, т.е. от 0 до 10 пойманных рыб*
$$P(X \leq 10) = \displaystyle\sum_{i = 0}^{10} \dfrac{2^{i} e^{-2}}{i!}$$

```python
P2 = sum(poisson_proba(k=i, lambda_=lambda_) for i in range(11))
print(P2)
```

*чтобы найти вероятность того, что число пойманных рыб при 200 забросах удочки будет не менее 5 **и** не более 10 умножим найденные значения:*

```python
P1 * P2
```
